```python

import numpy as np
import matplotlib.pyplot as plt

apps = apps.drop_duplicates()
serie_top_generos = apps.groupby("Genres").count().nlargest(6, "App").App.reset_index().Genres
data = apps.Genres[apps.Genres.isin(serie_top_generos)]

plt.hist(data, bins=15, color='blue', alpha=0.7)
plt.title('Histograma Generos')
plt.xlabel('Generos')
plt.ylabel('Frecuencia')
plt.show()
```
